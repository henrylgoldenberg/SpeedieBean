```python
# Take the data and make it csv, drop empty rows, add any columns we may need in the future
# In this code, we are joining the two data sets
import pickle5 as pkl
import pandas as pd
def joinCSV():
  with open("/content/incident_NFD_XDSegID_BigDataClass.pkl", "rb") as f:
      object = pkl.load(f)

  df1 = pd.DataFrame(object)
  new_df1 = df1.dropna()
  new_df1.to_csv('incidents.csv')

  with open("/content/inrix_BigDataClass.pkl", "rb") as f:
      object = pkl.load(f)
  
  df2 = pd.DataFrame(object)
  new_df2 = df1.dropna()
  new_df2.to_csv('intrix.csv')

  
  # Join the two datasets
  # Adding a column for day v.s. night and season

  new_df2.rename(columns={'geometry':'road_geometry'})
  join_df = df1.join(df2.set_index('XDSegID'), on='XDSegID', lsuffix='_left')
  new_df = join_df.dropna()
  mylist = []
  for index, row in new_df.iterrows():
    if (row['time_local'].hour > 20):
      mylist.append(1)
    elif (row['time_local'].hour < 6):
      mylist.append(1)
    else:
      mylist.append(0)
  new_df['night_or_not'] = mylist

  mynextlist = []
  for index, row in new_df.iterrows():
    if (row['time_local'].month <= 2):
      mynextlist.append('winter')
    elif ((row['time_local'].month > 2) and (row['time_local'].month <= 5)):
      mynextlist.append('spring')
    elif ((row['time_local'].month > 5) and (row['time_local'].month <= 8)):
      mynextlist.append('summer')
    elif ((row['time_local'].month > 8) and (row['time_local'].month <= 11)):
      mynextlist.append('fall')
    else:
      mynextlist.append('winter')
  new_df['season'] = mynextlist
  new_df.to_csv('incidents_join.csv')

joinCSV()
```

```python
# Calculating the mean response time per day over the course of the data set
def meanResponseTimeAllDays():
    """
    :return: List
    """
    # Defining query string to select the correct data
    query_str = "SELECT mean(response_time_sec) AS mean_response_time_sec FROM incidents WHERE time > '2016-12-30' AND time < '2021-04-01' GROUP BY time(1d) FILL(none)"
    result = list(client.query(query_str).get_points('incidents'))
    answer = []
    # Create a list of all the daily averages
    for point in range(len(result)):
      answer.append(round(result[point]['mean_response_time_sec'], 2))

    # define the time range
    begin = datetime.datetime(2016, 12, 30)
    end = datetime.datetime(2021, 4, 1)
    step = datetime.timedelta(days=1) # set sampling step as 1 day

    # generate X and Y axises
    timesteps = []
    while begin < end:
        timesteps.append(begin.strftime('%Y-%m-%d %H:%M:%S'))
        begin += step

    # plot the figure
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=timesteps, y=answer, mode='lines', name='Mean Response Time'))
    fig.update_layout(title="Mean Response Time for Incidents")
    fig.update_xaxes(title="Date")
    fig.update_yaxes(title="Mean Response Time in Seconds")
    fig.show()
    pass
```
